## 1. Introduction

Despite the increasing trend towards online shopping, high-street stores can still be highly profitable. In particular, stores that sell sport clothes and accessories are still important because customers often require specific advice on products, which cannot be always obtained online, and because they may want to try the product before buying it. However, it can take a lot of time and efforts for store managers to gain knowledge about specific sports and learn about the best products to sell in order to be competitive. Therefore, if a store owner wants to open a new store in a new city, she/he should find a city with a demand similar to that in the city where he currently has a store so that she/he does not need to change its product range significantly and can maintain her/his knowledge relevant. 
In this project, I have clustered the 25 most populous cities in the UK based on their top ten sport venue categories, e.g., basketball, tennis, and volleyball courts, using k-means clustering. I have assumed that the demand for sport products follows the availability of sports venues; for example, if there are many tennis courts and zero football pitches in a city, there will probably be higher demand for tennis shoes than for football shoes. The results of this project would be useful for a sport shop owner to know in which city/cities there is a sport product demand similar to that/those in which he currently has stores.

## 2. Data

For this project, I started by scraping the webpage https://www.thegeographist.com/uk-cities-population-1000/ to obtain a dataframe with the 25 most populous cities in the UK and their population. Then, I obtained the latitude and longitude of each city using the geopy library (https://geopy.readthedocs.io/en/stable/). I characterized the sport trend in each city using the location technology platform Foursquare (https://foursquare.com/ ). I searched for the following categories within a radius of 10 km from the city center: badminton courts, baseball fields, basketball courts, golf courses, boxing gyms, climbing gyms, gymnastics gyms, martial arts dojos, pilates studios, yoga studios, hockey fields and rinks, rugby pitches, skate parks, skating rinks, soccer fields, squash courts, tennis courts, and volleyball courts. Then, I calculated the top ten categories within each city, and used this feature to cluster the cities.

## 3. Methodology

I assumed that the availability of specific sport venues in a city, indicates that people in the city favor the sport. In addition, this availability is assumed to create a demand. Therefore, a sportswear store in a city should have more demand for articles required to play the sports that are available in the city.
I used the k-means algorithm, which is an unsupervised clustering algorithm. The clustering algorithm was suitable because I did not have information on the sportswear market, but I needed to fin information within the data available in terms of the sport patterns in different cities. The k-means algorithm is a partitioning algorithm that clusters groups of objects so that the objects within a cluster are as similar as possible and objects in different clusters are as different as possible. 

### 3.1. Feature Selection

As a discriminating feature, I considered the ten most common sports venues present within the city. First, I obtained the top 100 sport venue categories within each city from foursquare. Some of the categories returned where similar, e.g., climbing gyms and rock climbing spots. For simplicity these categories where labeled as the same venue. Furthermore, some venue categories were irrelevant for the analysis, e.g., playgrounds, General Entertainment, and Hotel; these venue categories were deleted. As a result I obtained a dataset with 17 unique venue categories.
Second, I changed the variables in the dataframe to categorical variables, in order to perform the clustering, and sorted the dataframe with the top ten most frequent sport venues in each city. 
Finally, I grouped the cities into five clusters based on the top ten venues in each city.

## 4. Results

The clusters obtained were as follows. The first cluster contained Birmingham, Leicester, Coventry, Belfast, Kingston Upon Hull, Newcastle upon Tyne, Stoke on Trent, Southampton, and Northampton. In this cluster, soccer is the most common sport, followed by golf and martial arts. The second cluster contained Sheffield, Leeds, Edinburg, and Bradford; in this cluster golf is the most common sport, followed by soccer and martial arts. The third cluster contained only London, where Yoga seems to be the most common sport, followed by tennis and soccer. The fourth cluster contained Liverpool, Cardiff, Plymouth, and Reading, where golf and tennis were the most common sports, followed by soccer. The fifth cluster contained Glasgow, Bristol, Manchester, Nottingham, Derby, Portsmouth, and Brighton; here, the most common sports seem to be Soccer, followed by golf and martial arts. In cities in clusters one and five, the most common sports are the same, and the main difference seems to be the popularity of Yoga and Rugby in cluster five.
